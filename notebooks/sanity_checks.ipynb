# Sanity Checks
Run quick checks on a CSV dataset before full pipeline execution.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/raw/sample.csv')
if data_path.exists():
    df = pd.read_csv(data_path)
    print(df.shape)
    print(df.head())
    print(df.isna().sum().sort_values(ascending=False).head(10))
else:
    print('Add a dataset at data/raw/sample.csv to run this notebook cell.')
```
